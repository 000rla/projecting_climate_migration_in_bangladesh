```python
from ABM_model_steps import *
import time
import calendar
import numpy as np

# #define vars

mig_threshold = 1000 #migration threshold
mig_util = 400 #utility to migrate
ag_factor = 100 #land productivity factor
wealth_factor = 3000 #initialization of wealth factor
decision = ["utility",'utility_return_time','push_threshold','tpb','pmt','mobility_potential','hybrid'] #different decision methods avalible
comm_scale = 0.4 #proportion of community inpacted
shock_method='weather'
network_type=['random','small_world','preferential','fully_connected'] #different network types avalible
w1 = 0.15
w2 = 0.15
w3 = 0.7
k =10
threshold=20
scenario =[119,126,245,370,585] #different SSP climate senarios

#x=scenario[0]
y=decision[-1]
z=network_type[0]
#j=comm_scale[4]
no_months=12*30#(55-15)
start_time = time.time()
for i in range(1,5):
    print('run:',i)
    for x in scenario:
        print('starting scenario:',x)
        model = ABM_Model(y, mig_util, mig_threshold, wealth_factor, ag_factor, comm_scale,
                      shock_method, z, w1, w2, w3, k, threshold, x,
                     testing=False,factor=10000,binary=False)
        end_time = time.time()
        elapsed_time = end_time - start_time
        print('model initalised after',round(elapsed_time/60,2),'mins')
        print('---')

        #run model for ticks
        while (model.tick < no_months):
            model.model_step()
            model.data_collect()
            model.tick_up()

            time_values = model.step_time.values
            time_pd = pd.to_datetime(time_values)
            time_dt = np.array(time_pd, dtype='datetime64')
            year = np.datetime64(time_dt, 'Y').astype(int)+2000-30
            month = np.datetime64(time_dt, 'M').astype(int) % 12 + 1
            print(model.tick,':',calendar.month_name[month],year,'completed')
            print(model.mig_total,'migrants since Jan 2011')
            print('---')

        model.save_files(i)
        model.plotting()
        end_time = time.time()
        elapsed_time = end_time - start_time 
        print('Ran in',round(elapsed_time/60,2),'mins')
        print('---')
```
